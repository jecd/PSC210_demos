```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.getcwd()
```

```python
basedir = os.getcwd()
subdir = os.path.join(basedir,"ds000114", "sub-01", "ses-test", "func")
print(subdir)
```

```python
realignment = pd.read_table(subdir + "/sub-01_ses-test_task-fingerfootlips_bold_st_mc.nii.gz.par", sep="\s\s",names=[ "pitch", "roll", "yaw","x","y","z",])
framewise_rel = pd.read_table(subdir + "/sub-01_ses-test_task-fingerfootlips_bold_st_mc.nii.gz_rel.rms")
framewise_abs = pd.read_table(subdir + "/sub-01_ses-test_task-fingerfootlips_bold_st_mc.nii.gz_abs.rms")
```

```python
realignment
```

```python
framewise_rel
```

```python
framewise_abs
```

```python
# rotations pitch roll yaw (radians)
for i in range(0,3):
    plt.plot(realignment.iloc[:,i])
plt.xlabel("volumes")
plt.ylabel("rotation (radians)")
plt.title("pitch,roll,yaw in radians")
plt.show()

# translation (xyz) mm
for i in range(3,6):
    plt.plot(realignment.iloc[:,i])
plt.xlabel("volumes")
plt.ylabel("tranlations (mm)")
plt.title("x,y,z in mm")
plt.show()

## framewise displacment # in mm
plt.plot(framewise_rel)
plt.plot(framewise_abs)
plt.xlabel("volumes")
plt.ylabel("displacement (mm)")
plt.title("framewise displacmenent (mm)")
plt.show()
```
